```python
import os
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('book.csv', header=0)
books
```

**Omit some values of desired columns and row**

```python
books = books.drop('Unnamed: 0', axis = 1)
books = books.drop('Unnamed: 4', axis = 1)

books = books.drop(labels=[5,3,6], axis=0)
books
```

**Consider the missing values**

```python
books['Unnamed: 1'] = books['Unnamed: 1'].fillna("?")

books['Unnamed: 2'] = books['Unnamed: 2'].fillna(books['Unnamed: 2'].median())

books['Unnamed: 3'] = books['Unnamed: 3'].fillna(books['Unnamed: 3'].median())

books
```

**Convert the data in the correct form**

```python
books['Unnamed: 1'] = books['Unnamed: 1'].apply(lambda x: str(x).split('\n')[1] if x != "?" else x)

books['Unnamed: 2'] = books['Unnamed: 2'].astype('int')

books['Unnamed: 3'] = books['Unnamed: 3'].astype('float')

books
```

convert "no period" to the NaN of numpy because that's make the proccessing faster 

```python
books.replace("?", np.nan, inplace = True)
books
```
